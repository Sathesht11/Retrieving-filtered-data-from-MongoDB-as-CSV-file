```python
# append each document into a list
data=[]

for i in collection.find({'$and':[{'genres':{'$in':['Horror','Comedy','Thriller','Romance']}}
                                  ,{'awards.wins':{'$gte':1}},{'imdb.rating':{'$gte':7}},
                                  {'tomatoes.viewer.rating':{'$gte':3.3}}]},
                         {'_id':0,'title':1,'languages':1,'year':1}):
  data.append(i)
```

```python
# convert list of documents into DataFrame using Pandas

import pandas as pd

df = pd.DataFrame(data)
df
```
